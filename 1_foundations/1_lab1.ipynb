.env ‚Üí load API key ‚Üí create Gemini client ‚Üí
send question ‚Üí receive AI response ‚Üí
extract text ‚Üí print ‚Üí display formatted

```python
from openai import OpenAI
from dotenv import load_dotenv
import os
from IPython.display import Markdown, display

# Load environment variables
load_dotenv(override=True)

# Gemini API configuration
GEMINI_BASE_URL = "https://generativelanguage.googleapis.com/v1beta/openai/"
google_api_key = os.getenv("GOOGLE_API_KEY")

# Create Gemini client
gemini = OpenAI(
    api_key=google_api_key,
    base_url=GEMINI_BASE_URL
)

# Messages (your question)
messages = [
    {"role": "user", "content": "Explain what is 2+2 in simple terms."}
]

# Send request to Gemini
response = gemini.chat.completions.create(
    model="gemini-2.5-flash",
    messages=messages
)

# Extract answer
answer = response.choices[0].message.content

# Print normal output
print(answer)

# Display nicely formatted Markdown (for Jupyter/Colab)
display(Markdown(answer)) 
```

```python
# First create the messages:

messages = [{"role": "user", "content": "pick a business area that might be worth exploring for an Agentic AI opportunity"}]

# Then make the first call:

response = gemini.chat.completions.create(
    model="gemini-2.5-flash-preview-05-20",
    messages=messages
)

# Then read the business idea:

business_idea = response.choices[0].message.content

# And repeat! In the next message, include the business idea within the message
print(business_idea)
```

```python
display(Markdown(business_idea))
```

```python
messages = [{"role": "user", "content": f"present a pain-point in {business_idea} industry"}]

response = gemini.chat.completions.create(
    model="gemini-2.5-flash-preview-05-20",
    messages=messages
)

pain_point = response.choices[0].message.content
print(pain_point)
display(Markdown(pain_point))

```

```python
messages = [{"role": "user", "content": f"propose the Agentic AI solution for {pain_point}"}]

response = gemini.chat.completions.create(
    model="gemini-2.5-flash-preview-05-20",
    messages=messages
)
solution = response.choices[0].message.content
print(solution)
display(Markdown(solution))

```
